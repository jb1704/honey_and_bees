```python
# percent renovated states the relation of renovated colonies to the maximum colonies

# calculate values for percent renovated column
bees['Percent Renovated'] = (bees['Renovated Colonies']*100/bees['Maximum Colonies']).round()
bees
```

```python
bees.isna().sum()
```

```python
# add new column Percent Lost

bees['Percent Lost']= (bees['Lost Colonies']*100/bees['Maximum Colonies']).round(0)
```

```python
#add year column
bees["Year"]=bees["Date"].dt.year

#drop values of 2021, as only the years 2015-2020 are required
bees = bees.drop(bees[bees["Year"] == 2021].index)

#change Renovated Colonies to float
bees['Renovated Colonies'] = pd.to_numeric(bees['Renovated Colonies'], downcast='float', errors='coerce')

bees
```

```python
#group by state and year and calculate means for column 2 to 13
bees_mean=bees.groupby(["State", "Year"])['Starting Colonies', 'Maximum Colonies', 'Lost Colonies','Added Colonies', 'Renovated Colonies','Percent Renovated', 'Varroa Mites (Percent)', 'Other pests and parasites (Percent)', 'Diseases (percent)', 'Pesticides (percent)', 'Other (percent)', 'Unknown (percent)' ].mean()

#renaming columns
bees_mean= bees_mean.rename(columns={"Starting Colonies": "starting_colonies",
                           "Maximum Colonies": "maximum_colonies",
                           "Lost Colonies": "lost_colonies",
                           "Added Colonies": "added_colonies",
                           "Renovated Colonies": "renovated_colonies",
                           "Percent Renovated" : "percent_renovated",
                           "Varroa Mites (Percent)": "varroa_mites_percent",
                           "Other pests and parasites (Percent)": "other_pests_and_parasites",
                           "Diseases (percent)": "diseases_percent",
                           "Pesticides (percent)": "pesticides_percent",
                           "Other (percent)": "other_percent",
                           "Unknown (percent)": "unknown_percent",
                           "Percent Lost": "lost_percent",
                           })

bees_mean
```

```python
bees_mean = bees_mean.round(2)
```

```python
bees_mean[['starting_colonies', 'maximum_colonies', 'lost_colonies', 'added_colonies', 'renovated_colonies']] = bees_mean[['starting_colonies', 'maximum_colonies', 'lost_colonies', 'added_colonies', 'renovated_colonies']].round()
```

```python
### honey ####

# Converting Units
# translate pounds into kilogras

honey
```

```python
honey_kg=honey.copy() 
honey_kg.iloc[:,2]=honey.iloc[:,2] * 0.454 # pounds in kg
honey_kg.iloc[:,3:5]=honey.iloc[:,3:5]*1000 * 0.454 # 1000 pounds in kg
honey_kg.iloc[:,5]=honey.iloc[:,5] / 0.454 
honey_kg=honey_kg.rename(columns={"Honey producing colonies (thousand)" : "producing_colonies_thousand", 
                                  "Yield per colony (pounds)" : "yield_per_colony_kg" , 
                                  "Production (1,000 pounds)" : "production_kg", 
                                  "Stocks December 15 (1,000 pounds)" : "stocks_15_12_kg",
                                  "Average price per pound (dollars)" : "avg_price_per_kg",
                                  "Value of production (1,000 dollars)" : "production_value_1000$"
                                  })
honey_kg
```

```python
# rounding

honey_kg.yield_per_colony_kg = honey_kg.yield_per_colony_kg.round(2)

honey_kg.avg_price_per_kg = honey_kg.avg_price_per_kg.round(2)
```

```python
honey_kg
```

```python
#drop United States rows
honey_kg = honey_kg.loc[~(honey_kg['State'].isin(['United States']))]

#drop rows called Other
honey_kg = honey_kg.loc[~(honey_kg['State'].isin(['Other']))]
honey_kg
```

```python
### Join Honey and Bees Set ###

#apply inner merge
merge = pd.merge(honey_kg, bees_mean, how='inner', on=['State', 'Year'])
merge
```
